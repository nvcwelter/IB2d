## Criação de vetores 


Vetores são nada mais, nada menos do que *numpy arrays* de 1 dimensão. Eles podem ser criados através de *listas* de python, ou usando funções do próprio numpy, como:
- `np.zeros`, que cria uma array de zeros com as dimensões especificadas
- `np.ones`, que faz o mesmo porém com 1s
- `np.random`, um módulo para criar arrays com números aleatórios com as dimensões especificadas

Neste último módulo, é possível criar arrays com números de 0 a 1 distribuidos uniformemente (`np.random.rand`) ou escolher uma distribuição específica, como por exemplo a distribuição normal (`np.random.randn`).

# 2. Código vetorizado VS Laços de repetição

Uma das vantagens de se usar o numpy é a de que as operações vetoriais são muito otimizadas. Quando se trabalha com redes neurais, por exemplo, esse tipo de otimização pode ser essencial para a implementação de um projeto.

Vamos utilizar as seguintes matrizes para nossos exemplos:

```python
m0 = np.random.randn(100, 100)
m1 = np.random.randn(100, 100)
m0
```

Podemos, por exemplo, fazer a multiplicação entre essas duas matrizes. Primeiro, vamos utilizar um laço `for`, percorrer cada elemento das duas matrizes, e somar a multiplicação destes elementos. Utilizaremos a biblioteca `time` para contabilizar o tempo de execução do nosso algoritmo.

```python
import time

ini = time.time()

# não é exatamente a implementação da multiplicação de matrizes, mas tem as mesmas operações
resp = 0
for i in m0:
  for j in m1:
    resp += i*j

print(time.time() - ini)
```

Vamos, agora, aplicar esta mesma operação, porém utilizando a implementação vetorizada do numpy.

```python
ini = time.time()
m0@m1
print(time.time() - ini)
```

E se aumentarmos de 100 para 1000 elementos? Quanto será o aumento do tempo de execução de cada algoritmo?

```python
m0 = np.random.randn(1000, 1000)
m1 = np.random.randn(1000, 1000)
m0
```

```python
ini = time.time()
resp = 0
for i in m0:
  for j in m1:
    resp += i*j
print(time.time() - ini)
```

```python
ini = time.time()
m0@m1
print(time.time() - ini)
```

Na célula abaixo, simulamos valores crescentes para o tamanho das matrizes, e guardamos o tempo de execução de cada algoritmo numa lista.

```python
tempos_sem_np = []
tempos_com_np = []

for i in range(3, 5):
  elem = 5**i
  m0 = np.random.randn(elem, elem)
  m1 = np.random.randn(elem, elem)

  ini = time.time()
  resp = 0
  for i in m0:
    for j in m1:
      resp += i*j
  tempos_sem_np.append(time.time() - ini)

  ini = time.time()
  m0@m1
  tempos_com_np.append(time.time() - ini)
```

Plotando os valores encontrados, vemos que o tempo de execução sem utilização de numpy cresce de forma mais rápida do que utilizando numpy (apesar de ambos os algoritmos apresentarem tempos de execução linear, ou seja, crescem proporcionalmente ao tamanho da entrada).

```python
import pandas as pd
t = pd.DataFrame([tempos_com_np, tempos_sem_np]).T
t.columns = ['com', 'sem']

t.plot()
```

# Exemplo de utilização: Sistema de busca


Vamos criar um sistema de busca para o conjunto de dados *Disaster tweets* (https://github.com/laxmimerit/twitter-disaster-prediction-dataset). Primeiro, montaremos a matriz abaixo, chamada de *bag of words* binário. Basicamente, cada coluna representa uma palavra do vocabulário (lista de palavras distintas do conjunto de dados), e cada linha representa um *tweet*. Se o tweet da linha i possui a palavra da coluna j, então o valor da posição i,j será 1. Caso contrário, será 0.:

image.png



Depois disso, precisaremos fazer este mesmo tratamento para nossa *query* de busca, ou seja, a frase que pesquisaremos no nosso sistema de busca. Se a query possuir a palavra na coluna j, então a posição j do vetor da query será igual a 1. Senão, será igual a 0.

image.png

Por fim, precisamos calcular um *score* para cada documento (ou seja, para cada tweet). Para isso, precisamos apenas fazer a soma da multiplicação de cada termo da query e do tweet em questão. O nosso sistema de busca retornará então os tweets ordenados por score, do maior para o menor (o *ranking* de busca).

image.png

A seguir, importamos o conjunto de dados utilizando o pandas.

```python
import pandas as pd
tweets = pd.read_csv('https://raw.githubusercontent.com/laxmimerit/twitter-disaster-prediction-dataset/master/train.csv')
tweets[['text']]
```

Utilizaremos o *CountVectorizer* do `sklearn`, que computa automaticamente o *bag of words*. Vemos que o dataset possui muitas palavras.

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(binary=True)
cv.fit(tweets['text'])
vocab = list(cv.vocabulary_.keys())
print(len(vocab), 'palavras')
```

Para ser mais fácil de visualizar, vamos trabalhar apenas com um subconjunto de 3 frases:

```python
cv_reduzido = CountVectorizer(binary=True)
cv_reduzido.fit(tweets['text'][10:13])
vocab = list(cv_reduzido.vocabulary_.keys())
print(len(vocab), 'palavras')
print(vocab)
```

Criaremos a matriz de *bag of words* binário utilizando um laço for.

```python
matriz_reduzida = []
for frase in tweets['text'][10:13]:
  matriz_reduzida.append([])
  for palavra in vocab:
    if palavra in frase: matriz_reduzida[-1].append(1)
    else: matriz_reduzida[-1].append(0)
print(np.array(matriz_reduzida))
```

Faremos o mesmo pré-processamento para a query "south tampa flooding".

```python
query_reduzida = 'south tampa flooding'
query_reduzida_vocab = [1 if palavra in query_reduzida else 0 for palavra in vocab]
print(query_reduzida_vocab)
```

Para cada linha da matriz, ou seja, para cada tweet, vamos calcular um score. No final, mostramos o ranking ordenado por score, do maior para o menor.

```python
s = []
for linha in matriz_reduzida:
  soma = 0
  for i in range(len(linha)):
    soma += linha[i]*query_reduzida_vocab[i]
  s.append(soma)

similaridade = {tweets['text'][10+i]: s[i] for i in range(3)}
ranking = {k: v for k, v in sorted(similaridade.items(), key=lambda item: item[1], reverse=True)}
ranking
```

Para vetorizar o código, podemos reescrever a operação que fizemos usando for como o produto escalar entre a query e cada linha da matriz.

```python
matriz_vec = np.array(matriz_reduzida)
query_vec = np.array(query_reduzida_vocab)
s = matriz_vec@query_vec
s
```

E temos o mesmo resultado!!! Yay!

Agora que nosso código está propriamente vetorizado, podemos aplicá-lo na matriz inteira. Vamos utilizar a função `fit_transform` do `CountVectorizer` para criar nossa matriz do *bag of words* binário, e vamos aplicar o métdodo `transform` na query para que seja feito o mesmo pré-processamento.

```python
query = 'south tampa flooding'

matriz = cv.fit_transform(tweets['text'])
query = cv.transform(pd.DataFrame([query])[0])
```

Vemos que o número de linhas e colunas das matrizes não permitiriam que fizéssemos o produto escalar entre elas. Por isso, precisaremos fazer a transposição da matriz do *bag of words*.

```python
print(query.shape, matriz.shape)
```

Finalmente, podemos aplicar o produto escalar na matriz completa. Imprimimos apenas os 10 primeiros colocados no ranking de busca.

```python
s = (query@matriz.T).toarray()[0]
similaridade = {tweets['text'][i]: s[i] for i in range(len(tweets))}
ranking = {k: v for k, v in sorted(similaridade.items(), key=lambda item: item[1], reverse=True)}
for i in list(ranking.items())[:10]:
  print(i)
```

Podemos juntar tudo isso numa única função. Trocamos, também, o *bag of words* binário pelo *Tfidf*, que é basicamente um *bag of words* mais rebuscado que leva em conta o número de tweets em que uma mesma palavra aparece (diminuindo, então, o peso das palavras "comuns" em todos os tweets).

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def search(query):
  cv = TfidfVectorizer()
  matriz = cv.fit_transform(tweets['text'])
  query = cv.transform(pd.DataFrame([query])[0])

  s = (query@matriz.T).toarray()[0]
  similaridade = {tweets['text'][i]: s[i] for i in range(len(tweets))}
  ranking = {k: v for k, v in sorted(similaridade.items(), key=lambda item: item[1], reverse=True)}
  for i in list(ranking.items())[:10]:
    print(i)
```

E agora podemos brincar com nosso próprio sistema de busca :)

```python
search('tornado')
```
